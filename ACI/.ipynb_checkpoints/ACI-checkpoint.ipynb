```python
# LISTA 01 - DSP
# Nome: David Machado Couto Bezerra & Gustavo Santiago
# Matricula: 2025128341
# Mestrado DEEC
import numpy as np
import matplotlib.pyplot as plt
import scipy
```

```python
fc1 = 1
fc2 = 2

t = np.linspace(0, 3, 10000)

s = np.sin(fc1*2*np.pi*t) + np.sin(fc2*2*np.pi*t)

plt.plot(t, s, 'b-', label=r'$s(t)$')
plt.xlabel('Time (s)')
plt.ylabel(r'$s(t)$')
plt.title('Continuous Signal')
plt.xlim(0, 3)
plt.legend()
plt.grid(True)
plt.show()
```
